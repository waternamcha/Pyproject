```python
from google.colab import files

# Upload the mnist_train.csv and mnist_test.csv files
uploaded = files.upload()

#1-a: Visualize the distribution of the 256 gray levels
import pandas as pd
import matplotlib.pyplot as plt

# Ensure plots are displayed in the notebook
%matplotlib inline

# Load the MNIST dataset
mnist_train = pd.read_csv('mnist_train.csv')

def visualize_gray_levels(data, num_images=5):
    for i in range(num_images):
        image = data.iloc[i, 1:].values.reshape(28, 28)
        plt.figure(figsize=(6, 4))
        plt.hist(image.ravel(), bins=256, range=[0, 256], density=True)
        plt.title(f'Histogram of pixel values for image {i}')
        plt.xlabel('Pixel value')
        plt.ylabel('Frequency')
        plt.show()

# Example usage
visualize_gray_levels(mnist_train, num_images=5)
```

```python
# 1-b: Display gray level values and represent them with color intensities
import seaborn as sns

def display_heatmap(data, num_images=5):
    for i in range(num_images):
        image = data.iloc[i, 1:].values.reshape(28, 28)
        plt.figure(figsize=(6, 6))
        sns.heatmap(image, cmap='gray', cbar=False)
        plt.title(f'Heatmap of image {i}')
        plt.show()

# Example usage
display_heatmap(mnist_train, num_images=5)
```

```python
# 1-c: Add a cross shape with a 2-pixel width to the specified image

def add_cross_to_image(data, image_index):
    image = data.iloc[image_index, 1:].values.reshape(28, 28)
    cross_shape = image.copy()
    center = 28 // 2
    cross_shape[center-1:center+1, :] = 255  # horizontal line
    cross_shape[:, center-1:center+1] = 255  # vertical line

    plt.figure(figsize=(6, 6))
    sns.heatmap(cross_shape, cmap='gray', cbar=False)
    plt.title(f'Image {image_index} with cross shape')
    plt.show()

# Example usage
add_cross_to_image(mnist_train, image_index=0)
add_cross_to_image(mnist_train, image_index=1)
add_cross_to_image(mnist_train, image_index=2)
add_cross_to_image(mnist_train, image_index=3)
add_cross_to_image(mnist_train, image_index=4)
```

```python
#Task 2b:

from sklearn.impute import SimpleImputer

# Define a SimpleImputer to fill NaN values with the mean of the column
imputer = SimpleImputer(strategy='mean')

# Apply the imputer to the data
data_imputed = imputer.fit_transform(data)

# Clamp values outside the range [0, 255]
data_clamped = np.clip(data_imputed, 0, 255)

# Convert back to DataFrame
df_cleaned = pd.DataFrame(data_clamped, columns=df.columns)

# Save the cleaned data to a new CSV file
df_cleaned.to_csv('MNIST_test_cleaned.csv', index=False)

print("Missing or unreasonable data handled and saved to 'MNIST_test_cleaned.csv'.")
```

```python
#Extra 1:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the given CSV file
df = pd.read_csv('MNIST_test_missing_data.csv')

# Convert the dataframe to a numpy array for easier manipulation
data = df.values

# Identify where the missing or unreasonable values are
invalid_indices = np.where((data[:, 1:] < 0) | (data[:, 1:] > 255) | np.isnan(data[:, 1:]))

# List incomplete images
incomplete_images = np.unique(invalid_indices[0])
print(f"Incomplete images (indices): {incomplete_images}")

# Function to visualize the invalid data points in the images
def visualize_invalid_data(image_data, invalid_idx, images_per_row=6):
    total_images = len(invalid_idx)
    rows = total_images // images_per_row + int(total_images % images_per_row != 0)

    fig, axes = plt.subplots(rows, images_per_row, figsize=(12, 2 * rows))

    for i, idx in enumerate(invalid_idx):
        image = image_data[idx, 1:].reshape(28, 28)  # Exclude the label column and reshape
        invalid_pixels = (image < 0) | (image > 255) | np.isnan(image)

        ax = axes[i // images_per_row, i % images_per_row]
        ax.imshow(image, cmap='gray')
        ax.set_title(f'Index: {idx}')
        ax.axis('off')

        # Overlay invalid data points
        for x in range(28):
            for y in range(28):
                if invalid_pixels[x, y]:
                    ax.scatter(y, x, c='red', s=1)  # Highlight invalid points in red

    # Remove any empty subplots
    for j in range(total_images, rows * images_per_row):
        fig.delaxes(axes.flatten()[j])

    plt.tight_layout()
    plt.show()

# Visualize the out-of-range and NaN data
visualize_invalid_data(data, incomplete_images)
```

Visulization of images with out of rage or missing value vs Cleaned data

```python
#Extra 2:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the original and cleaned CSV files
df_original = pd.read_csv('MNIST_test_missing_data.csv')
df_cleaned = pd.read_csv('MNIST_test_cleaned.csv')

# Convert the dataframes to numpy arrays for easier manipulation
data_original = df_original.values
data_cleaned = df_cleaned.values

# Identify where the missing or unreasonable values are in the original data
invalid_indices = np.where((data_original[:, 1:] < 0) | (data_original[:, 1:] > 255) | np.isnan(data_original[:, 1:]))

# List incomplete images
incomplete_images = np.unique(invalid_indices[0])
print(f"Incomplete images (indices): {incomplete_images}")

# Filter out indices that are out of bounds for the cleaned data
incomplete_images = incomplete_images[incomplete_images < len(data_cleaned)]
print(f"Filtered incomplete images (indices): {incomplete_images}")

# Function to visualize the comparison of original and cleaned images
def visualize_comparison(original_data, cleaned_data, invalid_idx, images_per_row=3):
    total_images = len(invalid_idx)
    rows = total_images // images_per_row + int(total_images % images_per_row != 0)

    fig, axes = plt.subplots(rows, images_per_row * 2, figsize=(12, 2 * rows))

    for i, idx in enumerate(invalid_idx):
        original_image = original_data[idx, 1:].reshape(28, 28)  # Exclude the label column and reshape
        cleaned_image = cleaned_data[idx, 1:].reshape(28, 28)  # Exclude the label column and reshape

        # Original image with invalid pixels highlighted
        invalid_pixels = (original_image < 0) | (original_image > 255) | np.isnan(original_image)

        ax_orig = axes[i // images_per_row, (i % images_per_row) * 2]
        ax_clean = axes[i // images_per_row, (i % images_per_row) * 2 + 1]

        ax_orig.imshow(original_image, cmap='gray')
        ax_orig.set_title(f'Original Index: {idx}')
        ax_orig.axis('off')

        # Overlay invalid data points
        for x in range(28):
            for y in range(28):
                if invalid_pixels[x, y]:
                    ax_orig.scatter(y, x, c='red', s=1)  # Highlight invalid points in red

        # Cleaned image
        ax_clean.imshow(cleaned_image, cmap='gray')
        ax_clean.set_title(f'Cleaned Index: {idx}')
        ax_clean.axis('off')

    # Remove any empty subplots
    for j in range(total_images * 2, rows * images_per_row * 2):
        fig.delaxes(axes.flatten()[j])

    plt.tight_layout()
    plt.show()

# Visualize the comparison of original and cleaned images
visualize_comparison(data_original, data_cleaned, incomplete_images)
```

# Task 3:

**Implement image classification using the K-Nearest Neighbors (KNN) algorithm or other algorithms. You can either code it step-by-step or utilize well-prepared libraries like scikit-learn. Refer to online resources and tutorials for guidance. Show the accuracy on the test data. **

```python
import numpy as np  # Import numpy for array manipulation and numerical computations
import matplotlib.pyplot as plt  # Import matplotlib for plotting and data visualization
from tensorflow.keras.datasets import mnist  # Import MNIST dataset from tensorflow.keras
import pandas as pd  # Import pandas for data analysis and manipulation
from sklearn.model_selection import GridSearchCV  # Import GridSearchCV from sklearn for hyperparameter tuning

# Load MNIST dataset (includes both training and test data)
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
def visualize_training_testing_set(training_set, testing_set):
    # Define the names of the sets
    set_name = ['Training set', 'Testing set']
    # Calculate the number of images in each set
    image_count = [len(training_set), len(testing_set)]

    # Define how much each wedge should be "exploded" (separated from the center)
    explode = (0.15, 0.15)

    # Define a helper function to format the labels on the pie chart
    def func(pct, all_values):
        # Calculate the absolute number of images for the percentage given
        absolute = int(np.round(pct/100.0 * np.sum(all_values)))
        # Return the formatted string with percentage and absolute number in thousands
        return f"{pct:.2f}%\n({absolute/1000:.1f}k)"

    # Create the pie chart
    wedges, texts, autotexts = plt.pie(
        image_count,  # The values for the pie chart
        labels=set_name,  # The labels for each wedge
        autopct=lambda pct: func(pct, image_count),  # Function to format the percentage text
        textprops=dict(color="w"),  # Set the color of the text to white
        pctdistance=0.6,  # The ratio of the center circle compared to the pie radius
        explode=explode  # Explode the wedges for better visibility
    )

    # Add a legend to the pie chart
    plt.legend(
        wedges,  # The handles (elements) in the legend
        set_name,  # The labels for the legend
        title="Dataset",  # The title of the legend
        loc="center left",  # The location of the legend
        bbox_to_anchor=(1, 0, 0.5, 1)  # The bounding box of the legend
    )

    # Add a circle at the center to make it look like a donut chart
    centre_circle = plt.Circle((0, 0), 0.25, fc='white')
    fig = plt.gcf()  # Get the current figure
    fig.gca().add_artist(centre_circle)  # Add the center circle to the plot
    plt.title('Number of Training and Testing Images Distribution')  # Set the title of the plot
    plt.show()  # Display the plot

# Call the function to visualize the training and testing set distribution
visualize_training_testing_set(train_labels, test_labels)
```

```python
def visualize_data_distribution(array_label, type_data):
    # Get unique labels and their respective counts from the array
    unique_labels, label_counts = np.unique(array_label, return_counts=True)

    # Create a bar plot with the unique labels on the x-axis and their counts on the y-axis
    plt.bar(unique_labels, label_counts)
    plt.xlabel('Label')  # Label the x-axis
    plt.ylabel('Count')  # Label the y-axis
    plt.title(f'Distribution of Labels in {type_data} set')  # Set the title of the plot
    plt.xticks(unique_labels)  # Set the labels on the x-axis to be the unique labels

    # Set the position for the text labels on the bars
    y_plot_text_position = label_counts.min() // 2

    # Add text labels on the bars to show the count of each label
    for index, value in enumerate(label_counts):
        plt.text(
            index,  # x-coordinate (position of the bar)
            y_plot_text_position,  # y-coordinate (height of the text)
            str(value),  # The text to display (the count of the label)
            ha='center',  # Horizontal alignment of the text
            color='white',  # Color of the text
            size=8  # Size of the text
        )

    # Display the plot
    plt.show()


# Visualize the distribution of labels in the training set
visualize_data_distribution(train_labels, "Training")
# Visualize the distribution of labels in the test set
visualize_data_distribution(test_labels, "Testing")
```

**Implement KNN algorithm**

```python
# Import necessary libraries for KNN and accuracy calculation
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Flatten the images for KNN input
# KNN requires a 2D array where each row is a sample and each column is a feature
train_images_flattened = train_images.reshape(train_images.shape[0], -1)
test_images_flattened = test_images.reshape(test_images.shape[0], -1)

# Initialize KNN model (using default parameters)
knn = KNeighborsClassifier()

# Train the KNN model on the training data
knn.fit(train_images_flattened, train_labels)

# Predict labels for the test data using the trained KNN model
test_predictions = knn.predict(test_images_flattened)

# Calculate the accuracy of the predictions
accuracy = accuracy_score(test_labels, test_predictions)

# Print the accuracy of the KNN classifier
print(f'KNN Classification Accuracy: {accuracy:.4f}')
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def visualize_confusion_matrix(test_labels, test_predictions):
    # Build the confusion matrix
    cf_matrix = confusion_matrix(test_labels, test_predictions)

    # Create a new figure and axis with specified size
    cm_fig, cm_ax = plt.subplots(figsize=(8, 8))

    # Create a ConfusionMatrixDisplay object with the confusion matrix
    cf_mat_disp = ConfusionMatrixDisplay(cf_matrix)

    # Plot the confusion matrix on the provided axis with a blue color map
    cf_mat_disp.plot(ax=cm_ax, cmap=plt.cm.Blues)

    # Display the plot
    plt.show()

# Call the function to visualize the confusion matrix for the test set predictions
visualize_confusion_matrix(test_labels, test_predictions)
```

**Improve the model performance**

```python
# Define the parameter grid for GridSearchCV
# We are testing different values for 'weights' and 'n_neighbors'
grid_search_param = [
    {"weights": ["uniform", "distance"],  # 'uniform': all points are weighted equally, 'distance': closer points are weighted more
     "n_neighbors": [2, 3, 5, 7]  # Number of neighbors to use in the KNN algorithm
    }
]

# Initialize GridSearchCV with the KNN model, parameter grid, accuracy as the scoring metric, 5-fold cross-validation, and verbose output
knn_grid_search = GridSearchCV(
    knn,  # The KNN model
    param_grid=grid_search_param,  # The parameter grid to search over
    scoring="accuracy",  # The metric to evaluate during the grid search
    cv=5,  # Number of folds for cross-validation
    verbose=3  # Verbosity level (higher number means more details in the output)
)

# Fit GridSearchCV on the flattened training images and labels
knn_grid_search.fit(train_images_flattened, train_labels)

# Print the best parameters found during the grid search
print(knn_grid_search.best_params_)
# Print the best accuracy score achieved with the best parameters
print(knn_grid_search.best_score_)
```

```python
# Convert the cross-validation results from the grid search to a pandas DataFrame
cv_results_df = pd.DataFrame.from_dict(knn_grid_search.cv_results_)

# Sort the DataFrame by the rank of the test scores
# 'rank_test_score' indicates the ranking of the models based on their performance (1 is the best rank)
sorted_cv_results_df = cv_results_df.sort_values(by='rank_test_score')

# Display the sorted DataFrame
sorted_cv_results_df
```

```python
# Retrieve the best estimator (model) found by GridSearchCV
knn_fine_tuned = knn_grid_search.best_estimator_

# Train the fine-tuned KNN model on the training data
knn_fine_tuned.fit(train_images_flattened, train_labels)

# Predict labels for the test data using the fine-tuned KNN model
test_predictions = knn_fine_tuned.predict(test_images_flattened)

# Calculate the accuracy of the predictions
accuracy = accuracy_score(test_labels, test_predictions)

# Print the accuracy of the fine-tuned KNN classifier
print(f'KNN Classification Accuracy: {accuracy:.4f}')
```

```python
# Visualize the confusion matrix for the fine-tuned KNN classifier
visualize_confusion_matrix(test_labels, test_predictions)
```
